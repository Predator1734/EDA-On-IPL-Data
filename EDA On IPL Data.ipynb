```python
#loading required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#to read csv file
ipl=pd.read_csv("matches.csv")
```

```python
#to display first 5 records of the data
ipl.head()
```

```python
#to get number of rows and columns from the data
ipl.shape
```

```python
#Getting frequency of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#Gettng the the top 10 of the most man of the match
ipl['player_of_match'].value_counts()[0:10]
```

```python
#Gettng the the top 5 of the most man of the match
ipl['player_of_match'].value_counts()[0:5]
```

```python
#making a bar-plot for the top 5 players with most man of the match
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="g")
plt.show()
```

```python
#Getting frequency of result column
ipl['result'].value_counts()
```

```python
#Getting out the number of toss wins wrt to each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records in which a team won the match batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#top 5 records of batting_first
batting_first.head()
```

```python
#making a histogram
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution Of Runs")
plt.xlabel("Runs")
plt.ylabel("Number Of Matches")
plt.show()
```

```python
#Finding out the number of wins of eac team
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
#making pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#making pie chart without percentage values
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()))
plt.show()
```

```python
#extracting the records where team won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
#making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
